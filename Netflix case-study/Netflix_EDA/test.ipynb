```python
# SQL query gets sent to MySQL database
query = "SELECT show_id, country FROM netflix_raw WHERE country IS NOT NULL"

# pandas sends this query to MySQL and gets results back
df3 = pd.read_sql(query, engine)
df3 
df3['country'] = df3['country'].str.split(',')
df_country = df3.explode('country')
df_country['country'] = df_country['country'].str.strip()
df_country = df_country[df_country['country'] != '']
df_country
df_country.to_sql('netflix_country', engine, if_exists='replace', index=False)
```

```python
# SQL query gets sent to MySQL database
query = "SELECT show_id, listed_in FROM netflix_raw WHERE listed_in IS NOT NULL"

# pandas sends this query to MySQL and gets results back
df4 = pd.read_sql(query, engine)
df4
df4['listed_in'] = df4['listed_in'].str.split(',')
df_listed_in = df4.explode('listed_in')
df_listed_in['listed_in'] = df_listed_in['listed_in'].str.strip()
df_listed_in = df_listed_in[df_listed_in['listed_in'] != '']
df_listed_in
df_listed_in.to_sql('netflix_genre', engine, if_exists='replace', index=False)
```

```python
# SQL query gets sent to MySQL database
query = "SELECT show_id, cast FROM netflix_raw WHERE cast IS NOT NULL"

# pandas sends this query to MySQL and gets results back
df5 = pd.read_sql(query, engine)
df5
df5['cast'] = df5['cast'].str.split(',')
df_cast = df5.explode('cast')
df_cast['cast'] = df_cast['cast'].str.strip()
df_cast = df_cast[df_cast['cast'] != '']
df_cast
df_cast.to_sql('netflix_cast', engine, if_exists='replace', index=False)
```
